```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Dados do Parlamentar

```python
df_meps = pd.read_csv("./data/silver/mep_membership_timeline.csv")

df_meps.fillna(0, inplace=True)

df_meps['member_id'] = df_meps['ID'].apply(lambda x: f"person/{x}")

del df_meps['ID']
```

```python
pd.set_option('display.max_rows', 80)
df_meps.describe().T.sort_index()
```

# Criar dummies

```python
cat_vars = ['POLITICAL_GROUP', 'COUNTRY']

df_dummies = pd.get_dummies(df_meps, columns=cat_vars, dtype=int)
```

```python
pd.set_option('display.max_rows', 150)
df_descriptives = df_dummies.describe().T.sort_index()
df_descriptives
```

# Agregar
usando max()

```python
from datetime import datetime

df_dummies['Y-m'] = df_dummies['date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d').strftime('%Y-%m'))
df_dummies['Y-w'] = df_dummies['date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d').strftime('%Y-W%W'))
```

```python
df_ym = df_dummies.copy()
del df_ym['date']
df_ym = df_dummies.groupby(['member_id', 'Y-m']).max().reset_index()


df_yw = df_dummies.copy()
del df_yw['date']
df_yw = df_dummies.groupby(['member_id', 'Y-w']).max().reset_index()
```

```python
df_ym.describe().T.sort_index()
```

```python
df_dummies.to_csv("./data/gold/panel_data_meps_v020702.csv", index=False)
df_ym.to_csv("./data/gold/panel_data_meps_Ym_v020702.csv", index=False)
df_yw.to_csv("./data/gold/panel_data_meps_Yw_v020702.csv", index=False)
```
