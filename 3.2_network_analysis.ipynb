```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_combined = pd.read_csv('./data/silver/df_matched.csv')
```

# Network analysis

Let's create a graph containing MEPs and entities as nodes.

The meetings the had with each other are the edges.

```python

# Create a bipartite graph
G = nx.Graph()

# Add nodes for organizations and MEPs
organizations = df_combined['Name'].unique()
meps = df_combined['member_name'].unique()

# Add nodes with different colors for organizations and MEPs
G.add_nodes_from(organizations, bipartite=0, node_color='lightblue')
G.add_nodes_from(meps, bipartite=1, node_color='lightgreen')

# Add edges between organizations and MEPs
for _, row in df_combined.iterrows():
    G.add_edge(row['Name'], row['member_name'])

# Set up the plot
plt.figure(figsize=(15, 15))
pos = nx.spring_layout(G, k=0.15, iterations=20)

# Draw the graph
nx.draw_networkx_nodes(G, pos, nodelist=organizations, node_color='lightblue', node_size=100)
nx.draw_networkx_nodes(G, pos, nodelist=meps, node_color='lightgreen', node_size=100)
nx.draw_networkx_edges(G, pos, alpha=0.2)
nx.draw_networkx_labels(G, pos, font_size=6)

plt.title('Network of Organizations and MEPs')
plt.axis('off')
plt.savefig('network_visualization.png', dpi=300, bbox_inches='tight')
plt.close()

# Calculate and print some basic network statistics
print(f"Number of organizations: {len(organizations)}")
print(f"Number of MEPs: {len(meps)}")
print(f"Number of connections: {G.number_of_edges()}")
print(f"Average degree: {sum(dict(G.degree()).values()) / G.number_of_nodes():.2f}") 
```
