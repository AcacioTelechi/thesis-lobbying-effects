```python
import pandas as pd
```

```python
df_graph_data = pd.read_csv("./data/gold/gephi_graph_data.csv")
```

```python
df_graph_data.head()
```

```python
df_graph_data.groupby("modularity_class").agg(
    {"Id": "count", "degree": "mean", "Authority": "mean", "eigencentrality": "mean"}
).reset_index()
```

```python
## A dataframe to get the top ten by authority by cluster

df_graph_data.groupby("modularity_class").apply(
    lambda x: x.sort_values("Hub", ascending=False).head(10)
).reset_index(drop=True).to_csv("./data/gold/top_ten_by_hub.csv", index=False)
```
