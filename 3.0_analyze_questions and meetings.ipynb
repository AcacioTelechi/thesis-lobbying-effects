```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_questions = pd.read_csv('./data/silver/df_questions_by_period.csv', index_col=0)
df_meetings = pd.read_csv('./data/silver/df_meetings_by_period.csv', index_col=0)

df_questions_ymd = pd.read_csv('./data/silver/df_questions_by_period_YYYY-MM-DD.csv', index_col=0)
df_meetings_ymd = pd.read_csv('./data/silver/df_meetings_by_period_YYYY-MM-DD.csv', index_col=0)
```

```python
# Get all unique periods from both dataframes
all_periods = sorted(set(df_questions.columns.tolist() + df_meetings.columns.tolist()))


# Fill missing periods with 0
df_questions = df_questions.reindex(all_periods, axis=1, fill_value=0)
df_meetings = df_meetings.reindex(all_periods, axis=1, fill_value=0)

# Get the column names (periods)
periods = df_questions.columns

# Calculate total questions and meetings per period
questions_by_period = df_questions[periods].sum()
meetings_by_period = df_meetings[periods].sum()

# Create the plot
plt.figure(figsize=(15, 7))
plt.plot(periods, questions_by_period, color='red', label='Questions')
plt.plot(periods, meetings_by_period, color='blue', label='Meetings')

# Customize the plot
plt.title('Questions and Meetings over Time')
plt.xlabel('Period')
plt.ylabel('Number of Questions and Meetings')
plt.legend(loc='upper right')

# Rotate x-axis labels for better readability and show fewer ticks
plt.xticks(range(0, len(periods), 6), periods[::6], rotation=45)

# Adjust layout to prevent label cutoff
plt.tight_layout()

plt.show()
```
